```python
import pandas as pd
```

```python
def get_mta_data(dates):
    file_path_string = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt'
    dfs = []
    for date in dates:
        df = pd.read_csv(file_path_string.format(date))
        dfs.append(df)
    return pd.concat(dfs)
```
